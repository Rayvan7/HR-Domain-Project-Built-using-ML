# Pipeline

```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline #####*************************
from sklearn.compose import ColumnTransformer # one-hot encoder gives dummy value where this columntransform will fit value
from sklearn.preprocessing import OneHotEncoder, StandardScaler 
from sklearn.impute import SimpleImputer, KNNImputer # missing vale handling
from imblearn.over_sampling import SMOTE # imbalance dataset
from imblearn.pipeline import Pipeline as ImbPipeline ####**********************
from sklearn.metrics import classification_report, accuracy_score
from sklearn.base import BaseEstimator, TransformerMixin ####*******************
```

```python
# preprocessing approach
# handling missing value, outlier treatement, imbalance treamtent, encoding, feature scaling
```

```python
# this is the code for outlier for any dataset where we can use this to find every outlier at one go

class OutlierClipper(BaseEstimator, TransformerMixin): # BaseEstimator:it provide a consistant interface means each and every
    def __init__ (self, factor=1.5):                   # variable, they will go one by one, and each and every variable they 
        self.factor = factor                           # will calculate the negative outiler and positive outlier
         
                                                       # TransformerMixin: to transform the value by capping method
    
    def fit(self, x, y = None): # y = None because we don't touch dependant variable for outliers
        x_df = pd.DataFrame(x)
        self.lower_bound = x_df.quantile(0.25) - (self.factor * (x_df.quantile(0.75) - x_df.quantile(0.25))) # Q1 - 1.5 * IQR
        self.upper_bound = x_df.quantile(0.25) + (self.factor * (x_df.quantile(0.75) - x_df.quantile(0.25))) # Q3 + 1.5 * IQR
        return self 
    def transform(self, x, y = None):
        x_df = pd.DataFrame(x)
        x_clipped = x_df.clip(lower=self.lower_bound, upper=self.upper_bound, axis=1) # axis = 1 is column wise
        return x_clipped.values

    
# Load the data
file_path = 'Attrition.csv'
data = pd.read_csv(file_path)


# split the data into independant variable and dependant variable
x = data.drop('Attrition', axis=1) # axis = 1 is column wise
y = data['Attrition']


# split the data into train and test 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


# preprocessing 
# imputation method - categorical and numerical variable (categorical - mode value , numerical - mean or median value), 
# if there is a missing data we go by these (sugession :- go by median value not mean value if it is numerical value
#                                                        and for categorical value go with mode - the most frequent value)

numerical_cols = x.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = x.select_dtypes(include=['object']).columns


# creating transformers for numerical and categorical feature

numerical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), # at one go we did SimpleImputer
                                        ('outlier_clipper', OutlierClipper()), # at one go we did OutlierClipper
                                        ('scaler', StandardScaler())]) # at one go we did StandardScaler

categorical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')), # use shift+tab and check
                                          ('onehot', OneHotEncoder(handle_unknown='ignore'))]) 


# create preprocessor using ColumnTransformer

preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, numerical_cols),
                                               ('cat', categorical_transformer, categorical_cols)])


# Define the pipeline with smote technique

pipeline = ImbPipeline(steps=[('preprocessor', preprocessor), ('smote', SMOTE(random_state=42)),
                              ('classifier', RandomForestClassifier(random_state=42))])


# Train the Model
pipeline.fit(x_train, y_train)

# make prediction
y_pred = pipeline.predict(x_test)

# Evaluate  the model 
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
